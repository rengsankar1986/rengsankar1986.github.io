```python
import requests
import pandas as pd
import lxml.html as LH

from bs4 import BeautifulSoup
```

```python
# Request content from the art collection archive
# For this program, we are scraping the first page of artist names starting with R
page = requests.get('https://web.archive.org/web/20121007172941/http://www.nga.gov/collection/anR1.htm')

# Create soup object
soup = BeautifulSoup(page.text, 'html.parser')

last_links = soup.find(class_='AlphaNav')
last_links.decompose()

artist_name_list = soup.find(class_='BodyText')
artist_name_list_items = artist_name_list.find_all('a')

# Use .contents to pull out the <a> tag’s children
for artist_name in artist_name_list_items:
    names = artist_name.contents[0]
    print(names)
```

```python
name = []
link = []

# Append the artist names and their links to two lists
for artist_name in artist_name_list_items:
    names = artist_name.contents[0]
    links = 'https://web.archive.org' + artist_name.get('href')
    name.append(names)
    link.append(links)
    print(names)
    print(links)
```

```python
# Combine the lists to a data frame
artist_info = list(zip(name,link))
artist_df = pd.DataFrame(artist_info, columns=['Name','Link'])
artist_df
```
